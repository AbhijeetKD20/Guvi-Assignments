Here I made a function in-order to find out different plots on basis of the features

```python
def diagnostic_plots(df, variable, target):
    #The function takes a dataframe "dataset"
    plt.figure(figsize = (20, 4))
    
    
    #histogram
    plt.subplot(1,4,1)
    sns.histplot(df[variable], bins = 30, color = "r")
    plt.title("Histogram")
    
    #scatterplot
    plt.subplot(1,4,2)
    plt.scatter(dataset[variable], df[target], color = "c")
    plt.title("Scatterplot")
    
    #boxplot
    plt.subplot(1,4,3)
    sns.boxplot(y = df[variable], color = "y")
    plt.title("Boxplot")
    
    #barplot
    plt.subplot(1,4,4)
    sns.barplot(x = target, y = variable, data = df)
    plt.title("Barplot")
    
    
    plt.show()
```

## **Splitting and Scaling the dataset**

```python
# #train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## **Logistic Regression** 

```python
#importing the Logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()

classifier.fit(X_train, y_train)
```

```python
print(dataset.cp.value_counts())
print("-----------------------")
print(dataset.restecg.value_counts())
print("-----------------------")
print(dataset.exng.value_counts())
print("-----------------------")
print(dataset.slp.value_counts())
print("-----------------------")
```

```python
#predicting the X_test from our Logistic regression model:
y_pred = classifier.predict(X_test)
```

**Model Performances**

```python
#Checking the accuracy and confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print("Confusion_Matrix")
print(cm)
print("Accuracy:", acc)
```

```python
#Classification_report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#Plotting the confusion_matrix
from sklearn.metrics import plot_confusion_matrix
print(plot_confusion_matrix(classifier,X_test,y_test))
```

## **K-NN Classifier**

Choosing the best K(neighbor) Value using cross validation score.

```python
from sklearn.neighbors import KNeighborsClassifier
```

Taking "k" till 25 values

```python
from sklearn.model_selection import cross_val_score
```

```python
knn_values = np.arange(1, 25)
cross_val_knn = []
for k in knn_values:
  knn_classifier = KNeighborsClassifier(n_neighbors= k)
  knn_classifier.fit(X_train, y_train)
  print("K value :", k, "train_score :", knn_classifier.score(X_train, y_train), "cross_val_score:", cross_val_score(knn_classifier, X_train, y_train).mean())
  cross_val_knn.append(cross_val_score(knn_classifier,X_train, y_train, cv = 10).mean())
```

```python
cross_val_knn_classifier = max(cross_val_knn)
print('The best K-Value is 12 and Cross_val_score is',cross_val_knn_classifier)
```

k=12 has the best score

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 12)
knn_classifier.fit(X_train, y_train)
```

```python
cross_val_knn_classifier= cross_val_score(knn_classifier, X_train, y_train, cv = 12).mean()
cross_val_knn_classifier
```

```python
#predicting the X_test from our Logistic regression model:
y_pred = knn_classifier.predict(X_test)
```

### **Model Performances**

```python
#Checking the accuracy and confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print("Confusion_Matrix")
print(cm)
print("Accuracy:", acc)
```

```python
#Classification_report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#Plotting the confusion_matrix
from sklearn.metrics import plot_confusion_matrix
print(plot_confusion_matrix(classifier,X_test,y_test))
```

## **Decision Tree Classifier**

### **Choosing the best of depth value**

Initially taking all values till 25.

```python
from sklearn.tree import DecisionTreeClassifier

max_depth = np.arange(1,15)
cross_val_dt = []
for d in max_depth:
  dt_classifier = DecisionTreeClassifier(max_depth = d, random_state = 0)
  dt_classifier.fit(X_train, y_train)
  print("Depth :", d, "train_score : ", dt_classifier.score(X_train, y_train), "cross_val_score :", cross_val_score(dt_classifier, X_train, y_train).mean()) 
  cross_val_dt.append(cross_val_score(dt_classifier, X_train, y_train, cv = 3).mean())
```

```python
cross_val_dt_classifier = max(cross_val_dt)
print("The best max_depth is 3 and Cross_val_score is", cross_val_dt_classifier)
```

Implementing Decision Tree Regression with max_depth = 3

```python
dt_classifier = DecisionTreeClassifier(max_depth = 3, random_state = 3)
dt_classifier.fit(X_train, y_train)
```

```python
cross_val_dt_classifier = cross_val_score(dt_classifier, X_train, y_train, cv=3).mean()
cross_val_dt_classifier
```

We can check the importance of the features: which feature is most/least important in our dataset in D.T.s

```python
Ft_imp = list(zip(dt_classifier.feature_importances_, dataset.columns[:-1]))
Ft_df = pd.DataFrame(Ft_imp, columns = ["Importance", "Feature"])
```
